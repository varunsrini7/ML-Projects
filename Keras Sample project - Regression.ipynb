```python
X = df.drop('price', axis = 1).values
y = df['price'].values
```

The use of .values is to ensure that we have numpy arrays for our model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

We only do transforming to the test set, whereas we do fitting and transforming to our train set. This is because we are working on the train data and we want to test it on our test data. If we do fitting to our test set then there will data leakage.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

We have 19 columns in our dataframe.  We keep the same number of neurons in our model, as there are features/size of the model.

```python
model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

#We are going kind of overkill with these many layer, so we
#will see how many such layers and epochs are required.
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
#Here mse is chosen as our data is linear(i.e price of houses)
#adam is chosen as its one of the best optimizers
```

```python
model.fit(x = X_train, y = y_train, validation_data = (X_test, y_test)
         , batch_size = 128, epochs = 400)
```

The use of validation data is, we can run our model.fit on our train data, and it takes some data from the training data and checks the loss on the test data after each epoch.A validation dataset is a sample of data held back from training your model that is used to give an estimate of model skill while tuning model's hyperparameters.  

The batch size is a hyperparameter of gradient descent that controls the number of training samples to work through before the model's internal parameters are updated. The number of epochs is a hyperparameter of gradient descent that controls the number of complete passes through the training dataset.

```python
model.history.history
```

```python
pd.DataFrame(model.history.history)
```

Now we have two losses, the loss on the training set and the validation loss on the test set

So we can directly compare the loss for validation and the loss on the training set by directly plotting. 

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

Now we can directly compare the value between the training loss and the validation loss and check for overfitting or underfitting of data.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
mean_squared_error(y_test, predictions)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df['price'].describe()
```

```python
5.402966e+05
```

So our error is around 20% of the mean, which isn't that great.

```python
explained_variance_score(y_test,predictions)
```

The explained variance score should be as close as possible to 1 to ensure that our model is accurate enough. 

```python
plt.figure(figsize = (12,6))
plt.scatter(y_test,predictions)
plt.plot(y_test, y_test, 'r')
```

The red line tells us the perfect prediction line[the 'r' just means that the color of the line should be red]
